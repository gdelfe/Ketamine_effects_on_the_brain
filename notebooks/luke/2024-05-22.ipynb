# Plot channel of cells discovered by spike sorting

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import src.paths
```

```python
regions = ['hpc', 'pfc']
dates = [
    '2022-07-27',
    '2022-07-28',
    '2022-08-01',
    '2022-08-03',
    '2022-08-04',
    '2022-08-06',
    '2022-08-08',
    '2022-08-11',
    '2022-08-12',
    '2022-08-13',
    '2022-08-15',
    '2022-09-12',
    '2022-09-14',
    '2022-09-16',
]
```

```python
for region in regions:
    fig, ax = plt.subplots(5, 3, figsize=(16, 16))
    plt.suptitle(f'{region.upper()} cell channels')
    for i in range(14):
        date = dates[i]
        raw = pd.read_pickle(src.paths.DATA / date / f'{date}-{region}-spikes.pkl')
        plt.sca(ax.flat[i])
        y = list(raw.columns.to_frame().channel)
        plt.plot(y)
        plt.title(f'Session {date}')
        plt.xlabel('Cell')
        plt.ylabel('Channel')
        plt.xlim([0, len(y) - 1])
        plt.ylim([0, 384])
    plt.tight_layout()
    plt.savefig(f'{region}-cell-channels.png', dpi=200)
    plt.show()
```
