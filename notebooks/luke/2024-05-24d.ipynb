```python
import src.paths
```

```python
sessions = {
    'M015': [
        '2022-07-27',
        '2022-08-01',
        '2022-08-03',
    ],
    'M016': [
        '2022-07-28',
        '2022-08-04',
        '2022-08-06',
    ],
    'M017': [
        '2022-08-08',
        '2022-08-12',
    ],
    'M018': [
        '2022-08-11',
        '2022-08-13',
        '2022-08-15',
    ],
    'M023': [
        '2022-09-12',
        '2022-09-14',
        '2022-09-16',
    ]
}
```

```python
def plot_unit_locations(indiv, region):
    fig, ax = plt.subplots(3, 1, figsize=(12, 8))
    plt.suptitle(f'{indiv} {region.upper()} unit locations')
    for i, date in enumerate(sessions[indiv]):
        df = pd.read_pickle(src.paths.DATA / date / f'{date}-hpc-spikes.pkl')
        units = df.columns.to_frame()
        plt.sca(ax.flat[i])
        sns.barplot(units.channel.value_counts().sort_index())
        plt.title(f'Session {date}')
        plt.xticks(rotation=90)
    plt.tight_layout()
    return plt.gcf()
```

```python
for region in ['hpc', 'pfc']:
    for indiv in sessions.keys():
        plot_unit_locations(indiv, region)
        plt.savefig(f'{indiv}-{region}-unit-depths.png')
        plt.show()
```
