```python
import os
from datetime import datetime
import pandas as pd
import numpy as np
from src import paths
from src.luke.load import load_metadata
```

```python
sessions = pd.read_csv(f'{paths.REF}/sessions.csv')
```

```python
from glob import glob
```

```python
def find_file(sess, signal='lfp', region='hpc', ftype='data'):
    date = sess.path[:10]
    probe = {'pfc': 'imec0', 'hpc': 'imec1'}[region.lower()]
    sig = {'lfp': 'lf', 'ap': 'ap'}[signal.lower()]
    ext = {'data': 'bin', 'meta': 'meta'}[ftype]
    pattern = str(paths.DATA / date / f'*.{probe}.{sig}.{ext}')
    matches = glob(pattern)
    if len(matches) == 0:
        raise ValueError(f'no matches found: {pattern}')
    return matches[0]
```

```python
metadata = {}
for region in ['hpc', 'pfc']:
    metadata[region] = {}
    for signal in ['lfp', 'ap']:
        l = []
        for i, sess in sessions.iterrows():
            metafile = find_file(sess, signal, region, 'meta')
            path = os.path.join(paths.DATA, metafile)
            md = load_metadata(path)
            md = pd.Series(md)[[
                'fileCreateTime', 'fileSizeBytes', 'fileTimeSecs', 'firstSample',
                'imAiRangeMax', 'imAiRangeMin', 
                'imMaxInt', 'imSampRate', 'imroTbl'
            ]]
            l.append(md)
        metadata[region][signal] = pd.DataFrame(l, index=sessions.index)
```

```python
df = []
for recid in range(36):
    for region in ['hpc', 'pfc']:
        for signal in ['lfp', 'ap']:
            md = metadata[region][signal].loc[recid]
            ser = {
                'recording': recid,
                'signal': signal,
                'region': region,
                'created': datetime.fromisoformat(md.fileCreateTime),
                'sizebytes': int(md.fileSizeBytes),
                'duration': float(md.fileTimeSecs),
                'firstsample': int(md.firstSample),
                'samplerate': int(md.imSampRate),
                'gain': int(md.imroTbl.strip('()').split(')(')[1:][0].split(' ')[{'ap': 3, 'lfp': 4}[signal]]),
                'minvolts': md.imAiRangeMin,
                'maxvolts': md.imAiRangeMax,
                'bitdepth': md.imMaxInt,
                'chanstart': 0,
                'chanend': 383,
                'chansys': 384,
                'nchans': 385,
            }
            df.append(ser)
df = pd.DataFrame(df)

cols = ['recording', 'signal', 'region']
df = df.sort_values(cols)
idx = df[cols]
df = df.drop(columns=cols)
idx = pd.MultiIndex.from_frame(idx)
df.index = idx
```
