```python
import pandas as pd
```

```python
root = '/Users/lukearend/phd/kfx'
df = pd.read_csv(f'{root}/ref/recid_CA1_DG_id_modified.csv', index_col=1)
df = df.drop(columns=[f'Unnamed: {i}' for i in (5, 9, 18, 21, 28)])
df = df.drop(columns=['Name_file', 'frame_used', 'fisure_old', 'notes'])
```

```python
cols = list(df.columns)
l = []
for s in cols:
    s = s.replace('Dent Up', 'dentup')
    s = s.replace('Dent Down', 'dentdown')
    s = s.replace(' ', '_')
    s = s.lower()
    l.append(s)
df.columns = l
df.index.name = 'recording'
```

```python
df.to_csv(f'{root}/ref/recordings.csv')
```
