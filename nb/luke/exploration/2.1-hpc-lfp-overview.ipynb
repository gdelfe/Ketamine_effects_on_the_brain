```python
import numpy as np
import pandas as pd

root = '/Users/lukearend/phd/kfx'
sessions = pd.read_csv(f'{root}/ref/sessions.csv')
recordings = pd.read_csv(f'{root}/ref/recordings.csv')
metadata = pd.read_csv(f'{root}/ref/metadata.csv')
```

```python
sessions.sort_values(['indiv', 'date'])
indivs = ['M015', 'M016', 'M017', 'M018', 'M023']
df1 = sessions[sessions.indiv.isin(indivs)].drop(columns='path')
```

```python
df2 = recordings[['recording', 'lfp_start', 'lfp_end', 'theta_start', 'theta_end']]
```

```python
hpc = metadata.region == 'hpc'
lfp = metadata.signal == 'lfp'
df3 = metadata[hpc & lfp].drop(columns=['region', 'signal', 'created', 'disk', 'duration'])
df = df1.merge(df2, on='recording').merge(df3, on='recording')
```

```python
df['int2volt'] = df.maxvolts / df.bitdepth
df = df.drop(columns=['maxvolts', 'minvolts', 'bitdepth', 'chansys', 'nchans'])
```
